```python
#2. Data Preprocessing
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
file_path = "Offenses_Data.xlsx"
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name="CustomQuery")
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("Missing Values in Each Column:")
print(missing_values)
```

```python
# Drop columns with excessive missing values
columns_to_drop = ['PLEA_FROM_OFFENDER','OFFENSE_COUNT','NATURE_OF_OFFENSE','SETTLEMENT_ACPT_DATE','DTM_OF_OFFENCE', 'TAF_ID', 'CASE_TYPE','REVENUE_LIABILITY_OF_GOODS', 'CASE_REFERENCE_NUMBER','OD_ID', 'CUSTOMS_STATION', 'SECTION_OF_LAW_CONTRAVENED', 'RECORDING_OFFENCE_DATE','SETTLEMENT_APPROVAL_DATE', 'OFFENSE_REFERENCE','DTM_OF_OFFENCE','VALUE_OF_GOODS','REVENUE_LIABILITY_OF_GOODS','STATEMENT_OF_OFFENSE','SETTLEMENT_REMARKS','OFFENDER_PIN','OFFENDER_NAME','COMPOUNDING_OFFICER_REMARKS']
df_cleaned = df.drop(columns=columns_to_drop, errors='ignore')
print(f"New dataset shape: {df_cleaned.shape}")
```

```python
# Fill missing numerical values with median
df_cleaned = df_cleaned.copy()
df_cleaned.loc[:, 'FINES_AND_PENALTIES'] = df_cleaned['FINES_AND_PENALTIES'].fillna(df_cleaned['FINES_AND_PENALTIES'].median())
df_cleaned.loc[:, 'AMOUNT_COLLECTED'] = df_cleaned['AMOUNT_COLLECTED'].fillna(df_cleaned['AMOUNT_COLLECTED'].median())
missing_after_fill = df_cleaned.isnull().sum().sum()
print(f"Number of missing values after filling: {missing_after_fill}")
```

```python
#Do categorical variables need encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_cleaned['CASE_STATUS'] = encoder.fit_transform(df_cleaned['CASE_STATUS'])
print("Unique values in CASE_STATUS after encoding:", df_cleaned['CASE_STATUS'].unique())
```
